```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

loading csv

```python
credit_card_data=pd.read_csv('/creditcard.csv')
```

```python
credit_card_data['Class'].value_counts()
```

dataset is highly imbalanced
0--> legit transation
0-->fraudlent transaction

```python
#seprating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

Under Sampling
building a sample dataset containg distribution of normal and fraudlent transaction
no of fraudlent transaction-->73

```python
legit_sample=legit.sample(n=73)
```

concatenating two data frame

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the data in to features and columns

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

splitting the data into training and testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model training

```python
model=LogisticRegression()
```

```python
#training the logistic model with the training data
model.fit(X_train,Y_train)
```

model evaluation

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('training data accuracy:',training_data_accuracy)
```

```python
#aacuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('test data accuracy:',test_data_accuracy)
```
